```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3,3,51)
```

```python
y1 = [i**2 for i in x]
y2 = [2*i + 0.5 for i in x]
y3 = [-3*i-1.5 for i in x]
y4 = [np.sin(i) for i in x]
```

```python
fig, ax = plt.subplots(2, 2)

ax[0, 0].set_title('График y1')
ax[0, 1].set_title('График y2')
ax[1, 0].set_title('График y3')
ax[1, 1].set_title('График y4')

ax[0, 0].set_xlim(-5,5)

ax[0, 0].plot(x, y1)
ax[0, 1].plot(x, y2)
ax[1, 0].plot(x, y3)
ax[1, 1].plot(x, y4)

plt.figure(1,[8,6])
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```
